```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

# LDA 적용하기

분류문제 차원축소에는 lda가 더 많이 적용됨

```python
lda = LinearDiscriminantAnalysis(n_components=2)
```

```python
#PCA와 달리 LDA는 클래스 정보도 필요
lda.fit(iris.data, iris.target) #feature, target
```

```python
iris_lda = lda.transform(iris.data)
iris_lda.shape
```

```python
import matplotlib.pyplot as plt

import seaborn as sns

import pandas as pd

iris_lda_df = pd.DataFrame(data=iris_lda ,columns=['component_1', 'component_2'])

iris_lda_df['target'] = iris.target
iris_lda_df.head()
```

```python
sns.scatterplot(x='component_1', y='component_2', hue='target', palette = 'muted', data=iris_lda_df)
plt.show()
```

```python
GMARKET_BEST_PRODUCT_URL = "https://corners.gmarket.co.kr/BestSellers"
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get(GMARKET_BEST_PRODUCT_URL)
```

```python
soup = BeautifulSoup(res.content, 'html.parser')
soup
```

```python
#nth-child 오류 나면 pip install beautifulsoup==4.8.1
```

```python
#best_item_sample = soup.select("#gBestWrap > div.gbest-top > div")[-1]
#best_item_sample.select("div > ul >li:nth-child(23)")
best_item_sample = soup.select_one("#gBestWrap > div.gbest-top > div > div.best-list > ul > li:nth-child(23)")
best_item_sample
```

```python
#상품명
best_item_sample.select_one("a.itemname").text
```

```python
#이미지 url 따기
best_item_sample.select_one("img.lazy")
```

```python
#src 가져오기
image_url = "https:"+ best_item_sample.select_one("img.lazy")['data-original']
print(image_url)
#이미지(로딩이 느린 이미지) : class 에 lazy 있으면 data-original 경로사용
```

```python
#사진저장
from urllib.request import urlretrieve

urlretrieve(image_url, "gmarket_image.png")
```

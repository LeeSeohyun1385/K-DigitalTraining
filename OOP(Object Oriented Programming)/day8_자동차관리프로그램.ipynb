```python
def 입력():
  global 리스트_최고속도
  global 리스트_최저속도
  이름=input("이름:")
  엔진가격=int(input("엔진가격:"))
  타이어가격=int(input("타이어가격:"))
  옵션=input("옵션(있다/없다):")
  if 옵션=='있다':
    옵션이름=input("옵션이름:")
  else:
    옵션이름='없음'
  최고속도=int(input("최고속도:"))
  리스트_최고속도 = 리스트_최고속도 if 리스트_최고속도>최고속도 else 최고속도
  리스트_최저속도 = 리스트_최저속도 if 리스트_최저속도<최고속도 else 최고속도
  자동차_정보=자동차정보(이름,엔진가격,타이어가격,옵션,옵션이름,최고속도)
  자동차_리스트.append(자동차_정보)
  print("차량 정보입력 완료")

def 목록출력():
  for 입력번호, 자동차 in [(i+1,v) for i,v in enumerate(자동차_리스트)]:
    print(f'{입력번호}번째 차량')
    print(f"""이름:{자동차.이름}
엔진가격:{자동차.엔진가격}
타이어가격:{자동차.타이어가격}
옵션:{자동차.옵션}
옵션이름:{자동차.옵션이름}
최고속도:{자동차.최고속도}
구매가격:{자동차.구매가격}""")
    print('-'*30) 
  print("저장된 목록 출력 완료")

def 가격출력():
  for 입력번호,자동차 in [(i+1,v) for i,v in enumerate(자동차_리스트)]:
    print(f'{입력번호}번째 차량') 
    print(f"""이름:{자동차.이름}
구매가격:{자동차.구매가격}""")
    print('-'*30)
  print("각 자동차 구매가격 출력 완료")

def 옵션_정보_출력():
  for 입력번호,자동차 in [t for t in [(i+1,v) for i,v in enumerate(자동차_리스트)] if t[1].옵션=='있다']:
    print(f'{입력번호}번째 차량') 
    print(f"""이름:{자동차.이름}
엔진가격:{자동차.엔진가격}
타이어가격:{자동차.타이어가격}
옵션:{자동차.옵션}
옵션이름:{자동차.옵션이름}
최고속도:{자동차.최고속도}
구매가격:{자동차.구매가격}""")
    print('-'*30)
  else:
    print("저장된 목록 출력 완료")

def 속도차_출력():
  print(f"가장 빠른차 속도 - 가장느린차 속도 = {리스트_최고속도-리스트_최저속도}")
  print("속도차 출력 완료")

def 종료():
  print("프로그램이 종료 됩니다.")
  return True
  
def UI():
  화면출력="""#### 자동차 정보 프로그램 ####
1. 자동차 정보 입력
2. 저장된 목록 보기
3. 각 자동차의 구매 가격 조회
4. 옵션이 있는 자동차 정보 조회
5. 빠른 자동차와 느린 자동차의 속도차 조회
6. 프로그램 이용 종료
입력:"""
  return input(화면출력)[0]
def 동작기():
  조건=UI()
  if 조건=='1':
    입력()
  elif 조건=='2':
    목록출력()
  elif 조건=='3':
    가격출력()
  elif 조건=='4':
    옵션_정보_출력()
  elif 조건=='5':
    속도차_출력()
  elif 조건=='6':
    return 종료()
  else:
    print("잘못입력되었습니다.\n입력은 1~6까지만 기능이있습니다.")
```

```python
def 입력():
  global 리스트_최고속도
  global 리스트_최저속도
  이름=input("이름:")
  엔진가격=int(input("엔진가격:"))
  타이어가격=int(input("타이어가격:"))
  옵션=input("옵션(있다/없다):")
  if 옵션=='있다':
    옵션이름=input("옵션이름:")
  else:
    옵션이름='없음'
  최고속도=int(input("최고속도:"))
  리스트_최고속도 = 리스트_최고속도 if 리스트_최고속도>최고속도 else 최고속도
  리스트_최저속도 = 리스트_최저속도 if 리스트_최저속도<최고속도 else 최고속도
  자동차_정보=자동차정보(이름,엔진가격,타이어가격,옵션,옵션이름,최고속도)
  자동차_리스트.append(자동차_정보)
  print("차량 정보입력 완료")

def 목록출력():
  for 입력번호, 자동차 in [(i+1,v) for i,v in enumerate(자동차_리스트)]:
    print('-'*30) 
    print(f'{입력번호}번|{자동차}')
    print('-'*30) 
  print("저장된 목록 출력 완료")

def 가격출력():
  for 입력번호,자동차 in [(i+1,v) for i,v in enumerate(자동차_리스트)]:
    print(f'{입력번호}번|{자동차.print_구매가격()}')
    print('-'*30)
  print("각 자동차 구매가격 출력 완료")

def 옵션_정보_출력():
  for 입력번호,자동차 in [t for t in [(i+1,v) for i,v in enumerate(자동차_리스트)] if t[1].옵션=='있다']:
    print(f'{입력번호}번|{자동차}')
    print('-'*30)
  else:
    print("저장된 목록 출력 완료")

def 속도차_출력():
  print(f"가장 빠른차 속도 - 가장느린차 속도 = {리스트_최고속도-리스트_최저속도}")
  print("속도차 출력 완료")

def 종료():
  print("프로그램이 종료 됩니다.")
  return True
  
def UI():
  화면출력="""#### 자동차 정보 프로그램 ####
1. 자동차 정보 입력
2. 저장된 목록 보기
3. 각 자동차의 구매 가격 조회
4. 옵션이 있는 자동차 정보 조회
5. 빠른 자동차와 느린 자동차의 속도차 조회
6. 프로그램 이용 종료
입력:"""
  return input(화면출력)[0]
```

클래스 기반 프로그래밍

```python
class 자동차정보:
  자동차_리스트=[]
  리스트_최고속도=0
  리스트_최저속도=501

  def __init__(self,이름,엔진가격,타이어가격,옵션,옵션이름,최고속도):
    self.이름=이름
    self.엔진가격=엔진가격
    self.타이어가격=타이어가격
    self.옵션=옵션
    self.옵션이름=옵션이름
    self.최고속도=최고속도
    self.구매가격=엔진가격+타이어가격

  def __str__(self):
    return f"이름:{self.이름}|엔진가격:{self.엔진가격}|타이어가격:{self.타이어가격}|옵션:{self.옵션}|옵션이름:{self.옵션이름}|최고속도:{self.최고속도}|구매가격:{self.구매가격}"
  def print_구매가격(self):
    return f"이름:{self.이름}|구매가격:{self.구매가격}"

  @classmethod
  def 입력(cls):
    이름=input("이름:")
    엔진가격=int(input("엔진가격:"))
    타이어가격=int(input("타이어가격:"))
    옵션=input("옵션(있다/없다):")
    if 옵션=='있다':
      옵션이름=input("옵션이름:")
    else:
      옵션이름='없음'
    최고속도=int(input("최고속도:"))
    cls.리스트_최고속도 = cls.리스트_최고속도 if cls.리스트_최고속도>최고속도 else 최고속도
    cls.리스트_최저속도 = cls.리스트_최저속도 if cls.리스트_최저속도<최고속도 else 최고속도
    자동차_정보=자동차정보(이름,엔진가격,타이어가격,옵션,옵션이름,최고속도)
    cls.자동차_리스트.append(자동차_정보)
    print("차량 정보입력 완료")

  @classmethod
  def 목록출력(cls):
    for 입력번호, 자동차 in [(i+1,v) for i,v in enumerate(cls.자동차_리스트)]:
      print('-'*30) 
      print(f'{입력번호}번|{자동차}')
      print('-'*30) 
    print("저장된 목록 출력 완료")

  @classmethod
  def 가격출력(cls):
    for 입력번호,자동차 in [(i+1,v) for i,v in enumerate(cls.자동차_리스트)]:
      print(f'{입력번호}번|{자동차.print_구매가격()}')
      print('-'*30)
    print("각 자동차 구매가격 출력 완료")

  @classmethod
  def 옵션_정보_출력(cls):
    for 입력번호,자동차 in [t for t in [(i+1,v) for i,v in enumerate(cls.자동차_리스트)] if t[1].옵션=='있다']:
      print(f'{입력번호}번|{자동차}')
      print('-'*30)
    else:
      print("저장된 목록 출력 완료")

  @classmethod
  def 속도차_출력(cls):
    print(f"가장 빠른차 속도 - 가장느린차 속도 = {cls.리스트_최고속도-cls.리스트_최저속도}")
    print("속도차 출력 완료")
  @classmethod
  def 종료(cls):
    print("프로그램이 종료 됩니다.")
    return True
  
  @classmethod
  def UI(cls):
    화면출력="""#### 자동차 정보 프로그램 ####
1. 자동차 정보 입력
2. 저장된 목록 보기
3. 각 자동차의 구매 가격 조회
4. 옵션이 있는 자동차 정보 조회
5. 빠른 자동차와 느린 자동차의 속도차 조회
6. 프로그램 이용 종료
입력:"""
    return input(화면출력)[0]
```

```python
동작리스트=["잘못입력되었습니다.\n입력은 1~6까지만 기능이있습니다.",자동차정보.입력,자동차정보.목록출력,자동차정보.가격출력,자동차정보.옵션_정보_출력,자동차정보.속도차_출력,자동차정보.종료]
run = False
while not run:
  n=자동차정보.UI()
  if n not in '123456':
    print(동작리스트[0])
  else:
    run=동작리스트[int(n)]()
```

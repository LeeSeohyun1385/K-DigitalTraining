레이블 인코딩

```python
#target에만 적용
from sklearn.preprocessing import LabelEncoder

items = ["청바지", "치마", "청바지", "원피스", "청바지", "치마", "치마", "청바지", "원피스"]

label_encoder = LabelEncoder()
label_encoder.fit(items)
```

```python
labels = label_encoder.transform(items)
labels
```

```python
#숫자를 다시 문자열로 디코딩
label_encoder.inverse_transform([2,1,0,0,1,2,0,0,1])
```

```python
#클래스 확인
label_encoder.classes_
```

원-핫인코딩(One-Hot Encoding)

```python
from sklearn.preprocessing import OneHotEncoder
ohe_encoder = OneHotEncoder() # -> Feature(행렬로 표현)
```

```python
#1차원 배열을 2차원 배열로 바꿔주기
# vector -> matrix
# N -> N X 1

import numpy as np

items_arr = np.array(items).reshape(-1,1)
items_arr
```

```python
ohe_labels = ohe_encoder.fit_transform(items_arr)
ohe_labels
```

```python
print(ohe_labels)
# 희소 행렬 (sparse matrix)
```

```python
ohe_labels.toarray()
```

Feature 스케일링

Min-Max Scaling(정규화)

```python
import pandas as pd
movie_score = pd.DataFrame({
    "naver":[0, 2, 4, 6, 8],
    "netflix":[1, 2, 3, 4, 5]
})

movie_score
```

```python
from sklearn.preprocessing import MinMaxScaler
 
min_max_scaler = MinMaxScaler()
```

```python
movie_score_scaled = min_max_scaler.fit_transform(movie_score)
movie_score_scaled
```

```python
pd.DataFrame(movie_score_scaled, columns = ["Naver Scaled", "Netflix Scaled"])
```

StandardScaling (표준화)
- 데이터를 원점 기준으로 정리  
- 평균 0, 표준편차 1 

```python
data1 = np.array([100, 110, 120, 130, 140])
data2 = np.array([0,1,2,3,4])
```

```python
data1.mean(), data1.std()
```

```python
data2.mean(), data2.std()
```

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
```

```python
data1_scaled = std_scaler.fit_transform(data1.reshape(-1,1))
data2_scaled = std_scaler.fit_transform(data2.reshape(-1,1))
```

```python
print(data1_scaled)
print(data2_scaled)
```

```python
data1_scaled.mean(), data1_scaled.std()
```

```python
data2_scaled.mean(), data2_scaled.std()
```

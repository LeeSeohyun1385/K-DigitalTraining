```python
import scrapy
import requests
from scrapy.http import TextResponse
```

# Xpath
- CSS Selector를 확장시킨 문법

```python
response = requests.get("http://corners.gmarket.co.kr/Bestsellers")
response.status_code
```

```python
response.url #응답한 url
```

```python
text_response = TextResponse(
    response.url,
    body = response.text,
    encoding = 'utf-8'
)
```

```python
text_response.xpath('//*[@id="itemidx1"]/a')
#반드시 ''(작은따옴표 사용)
```

```python
text_response.xpath('//*[@id="itemidx1"]/a/text()')
```

```python
text_response.xpath('//*[@id="itemidx1"]/a/text()').extract()
```

```python
type(text_response.xpath('//*[@id="itemidx1"]/a/text()').extract())
```

```python
text_response.xpath('//*[@id = "gBestWrap"]/div/div[3]/div/ul/li/a/text()').extract()
#//* 로 대부분 시작
#[]:속성
```

```python
text_response.xpath('//*[@id = "gBestWrap"]/div/div[3]/div/ul/li[1]/a/text()').extract()
#li에 숫자 바꿔가며 몇번째 뽑기 가능
```

```python
# 속성 추출은 속성명 앞에 @ 붙이기
text_response.xpath('//*[@id = "gBestWrap"]/div/div[3]/div/ul/li/a/@href').extract()
```

```python
text_response.xpath('//*[@id = "gBestWrap"]/div/div[3]/div/ul/li/div/a/img/@data-original').extract()
```

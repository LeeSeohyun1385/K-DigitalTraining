# Load Data

```python
from sklearn.datasets import load_iris

iris = load_iris()

iris_feature = iris.data
iris_target = iris.target

print("feature shape : {}".format(iris_feature.shape))
print("target shape : {}".format(iris_target.shape))
```

```python
import pandas as pd
feature_names = iris.feature_names
iris_df = pd.DataFrame(data = iris_feature, columns = feature_names)

iris_df["target"] = iris_target
iris_df.head()
```

```python
iris_df['target'].value_counts()
```

# 학습 데이터 / 테스트 데이터 분리

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    iris_df.drop("target", axis=1),
    iris_df["target"],
    test_size = 0.2,
    random_state = 42 #랜덤 시드 고정값
)

print(X_train.shape)
print(X_test.shape)
```

# 모델 훈련

```python
# 의사결정나무
from sklearn.tree import DecisionTreeClassifier

#모델 객체 생성
df_clf = DecisionTreeClassifier()

#모델 훈련
df_clf.fit(X_train, y_train)
```

# 테스트 데이터 예측

```python
pred = df_clf.predict(X_test) #예측은 feature만 넣기

print("테스트 세트의 예측 결과 : {}".format(pred))
```

# 평가

예측 정확도(Accuracy) 평가

```python
from sklearn.metrics import accuracy_score
acc_result = accuracy_score(y_test, pred)
print("예측 정확도 : {:.4f}".format(acc_result))
```

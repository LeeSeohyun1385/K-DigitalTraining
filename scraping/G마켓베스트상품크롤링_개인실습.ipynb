```python
import requests
from bs4 import BeautifulSoup
```

```python
GMARKET_BEST_PRODUCT_URL = "https://corners.gmarket.co.kr/BestSellers"
```

```python
res = requests.get(GMARKET_BEST_PRODUCT_URL)
```

```python
soup = BeautifulSoup(res.content, 'html.parser')
```

```python
best_item_sample = soup.select_one("#gBestWrap > div.gbest-top > div > div.best-list > ul> li ")
best_item_sample
```

설계 끝

```python
import re
product_list = []

best_item_sample = soup.select("#gBestWrap > div.gbest-top > div > div.best-list > ul> li ")


for item in best_item_sample:
  num = item.select_one("div.thumb > a")['onclick'].split()[5]
  code = re.sub(r'[^0-9]', '', num)
  page_url = item.select_one("div.thumb > a")['href']
  image_url = image_url = "https:"+ item.select_one("div.thumb > a > img.lazy")['data-original']
  price = item.select_one("div.item_price > div.s-price > strong > span").text
  product_list.append([code, page_url, image_url,price])

product_list
```

```python
import pandas as pd
best_item_df = pd.DataFrame(product_list)
best_item_df.columns = ["상품코드", "상세페이지url", "이미지url", "가격"]
best_item_df
```

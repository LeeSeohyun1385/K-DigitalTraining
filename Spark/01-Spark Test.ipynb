```python
from pyspark import SparkConf, SparkContext
```

```python
# 스파크 환경 설정 객체 생성
conf = SparkConf().setMaster("local").setAppName("country-student-counts")

# SparkContext 생성
# sc 변수명 고정
sc = SparkContext(conf=conf)
```

```python
!pwd
```

```python
filepath = '/home/ubuntu/working/spark-examples/data/xAPI-Edu-Data.csv'
```

```python
# 로컬 파일에서 데이터를 불러올 것이기 때문에 스킴이 file://
# 만약 HDFS에서 불러오려면? hdfs://

# local에 있는 데이터를 text파일로 불러오기 -> RDD가 됨
lines = sc.textFile(f"file:///{filepath}")
lines
```

```python
header = lines.first()
header
# Stage
```

```python
# 헤더를 제외한 데이터만 불러오기
datas = lines.filter(lambda row : row != header)
datas
```

```python
countries = datas.map(lambda row : row.split(",")[2])
countries

#datas.map() -> RDD.map(<task>)
#task : 작업 내용(함수)
# 스칼라, 파이썬 -> 함수형 프로그래밍 언어
# <task> : 코드조각이 worker한테 전달됨
```

```python
countries.collect() #Action. 모든 Transformations 작업을 한번에 수행 
```

```python
# 집계
# 각 나라별로 데이터가 몇개 있는지
result = countries.countByValue()
result
```

```python
# 작업이 끝났으면 반드시 스파크 컨텍스트를 종료
sc.stop()
```

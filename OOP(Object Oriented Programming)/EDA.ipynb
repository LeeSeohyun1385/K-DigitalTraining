### 결측치 처리
        1. 비우기
                의미없는 데이터 제거
                범주 데이터의 결측치는 가급적 제거
                과도한 결측치를 갖고있는 피처는 제거
                고유의 의미를 갖는 피처는 제거(단 주요피처가 아닐시)

        2. 채우기
                연속적인 데이터에 적용
                가급적 데이터 전반에 영향을 적게 주는 값으로 적용(ex평균등)
                외부 조간에 따라 조건을 반영한 내용을 적용

### 서수형,기수형데이터 인코딩
        1. 데이터 정형 
                컴퓨터가 이해할수있는 데이터 형으로 변경
                범주형을 숫자형으로 변경
                데이터 피처는 차원
        2. 범주형
                원 핫 인코딩

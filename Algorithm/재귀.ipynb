```python
## 함수
def openBox():
  global count
  print('상자를 엽니다.')
  count -= 1
  if (count == 0):
    print('** 선물 넣기 **')
    return
  openBox() #이 함수에 복사본이 있다고 생각
  print('상자를 닫습니다.')
## 메인
count = 10
openBox()
```

10~1의 합계 ....

```python
## 함수
def addNumber(num):
  if (num == 1):
    return 1
  return num + addNumber(num-1)
## 메인
print(addNumber(100))
```
